```python
# Importar bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset (descargado de Kaggle)
data = pd.read_csv('USA Housing Dataset.csv') 

# Mostrar las primeras filas
data.head()
```

```python
# Ver la información general del dataset
data.info()

# Descripción estadística del dataset
data.describe()

# Verificar valores faltantes
print(data.isnull().sum())
```

```python
# Visualización de la distribución de precios de casas
sns.histplot(data['price'], kde=True)
plt.title('Distribución de Precios de las Casas')
plt.show()
```

```python
# Eliminar columnas no numéricas que no serán utilizadas en la correlación
data_numeric = data.drop(columns=['date', 'street', 'city', 'statezip', 'country'])

# Ahora, generar la matriz de correlación con las columnas numéricas
plt.figure(figsize=(12, 8))
sns.heatmap(data_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación entre las Variables Numéricas')
plt.show()

```

```python
# Guardar los datos procesados para el siguiente notebook
data_numeric.to_csv('processed_house_data.csv', index=False)
print("Datos guardados para el siguiente notebook.")
```
